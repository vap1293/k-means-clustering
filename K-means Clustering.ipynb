#### K-means Clustering 

- Type of **clustering algorithm** used in **unsupervised learning**

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
#loading iris dataset

iris_dataset = load_iris()
```

```python
#checking the loaded dataset
#checking the infromation provided below, we have 4 features (sepal length, sepal width, petal length and petal width in cm/n)
#three types/classes Iris-Setosa, Iris-Versico and Iris-Virginica

print(iris_dataset)
```

```python
#slicing and storing dataset in other variable, 
#since the dataset contains a target variable as well, slicing of dataset is required to get only 4 features

sample_data = iris_dataset.data[:,:4]
```

```python
#verifying extracted data

print(sample_data)
```

```python
#visualizing sample_data for first two features without performing clustering

plt.scatter(x = sample_data[:,0],y = sample_data[:,1], marker='x', c = "blue",)
plt.title('Iris Dataset')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
#defining function to calculate distance (euclidean distance) between centroids and datapoints

from scipy.spatial import distance
def eucl_dist(xi,yi):
    dist = distance.euclidean(xi,yi)
    return dist
```

```python
#implementing K-Means Algortihm
#since we have a prior knowledge about dataset, wherein we have 3 types/classes of Iris flowers, we choose value of k = 3
#i.e. we will have 3 clusters representing each type Iris-Setosa, Iris-Versico and Iris-Virginica
#further using elbow method we verify that choosing value of k = 3 seems to be appropriate

#defining Kmeans class
class implemented_Kmeans:
    
    
    #defining constructor and setting some default parameter values
    def __init__(self, k=3, tol=0.001, iterations=100): 
        self.k = k
        self.tol = tol
        self.iterations = iterations
        
    #defining fit method
    def fit(self,dataset):
        
        #initailizng centroids
        self.centroids = {}        
        for i in range(self.k):
            self.centroids[i] = dataset[i]
            
        #initializing clusters
        for i in range(self.iterations):
            self.clusters = {}

            for i in range(self.k):
                self.clusters[i] = []

            #finding distance bewtween centroid and datapoint using defined function and checking minimum distance    
            for datapoint in dataset:
                euclidean_distance = [eucl_dist(datapoint,self.centroids[centroid]) for centroid in self.centroids]
                ind_cluster = euclidean_distance.index(min(euclidean_distance))
                self.clusters[ind_cluster].append(datapoint)

            previous_centroids = dict(self.centroids)

            for ind_cluster in self.clusters:
                self.centroids[ind_cluster] = np.mean(self.clusters[ind_cluster],axis=0)
                
            #setting boolean parameter which can be used to check if the centroids after iterative process have stabilised/converged  
            converged = True

            #iteratively updating centroids in clusters until converges or stabilizes
            for c in self.centroids:
                original_centroid = previous_centroids[c]
                updated_centroid = self.centroids[c]
                if np.sum((updated_centroid-original_centroid)/original_centroid*100.0) > self.tol:
                    converged = False

            if converged:
                break
```

```python
#instantiating class with default set parameters

model = implemented_Kmeans()
model.fit(sample_data)
```

```python
#visualizing/plotting the clustered dataset

colors = ["green","orange","yellow"]

for centroid in model.centroids:
    plt.scatter(model.centroids[centroid][0], model.centroids[centroid][1], marker="o", color="red")

for ind_cluster in model.clusters:
    color = colors[ind_cluster]
    for feature in model.clusters[ind_cluster]:
        plt.scatter(feature[0], feature[1], marker='x', color=color) #using first 2 features i.e. 2 dimension
plt.title('Clustered Iris Dataset using Implemented Kmeans Alg')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

```python
#using sklearn built-in Kmeans algortihm we can plot the elbow method on iris data to verify number of clusters (k) required
#additionaly using sklearn Kmeans algorithm and comparing/verifying ouptut with implemented Kmeans

from sklearn.cluster import KMeans
inertia = []

for i in range(1, 5):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 10, n_init = 10, random_state = 0) #instantiating
    kmeans.fit(sample_data) #using fit method of sklearn Kmeans
    inertia.append(kmeans.inertia_) #using inertia method of kmeans
    
#plotting above implementation to check for elbow

plt.plot(range(1, 5), inertia)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
#using sklearn Kmeans alg

kmeans_model = KMeans(n_clusters=3,max_iter=100,tol=0.001)
```

```python
#the fit_predict method will generate a target variable i.e. Lables (0,1,2) 

cluster = kmeans_model.fit_predict(sample_data)
```

```python
#plotting/visualizing clustered output of scikit learn Kmeans algorithm

plt.scatter(sample_data[cluster==0, 0], sample_data[cluster==0, 1], marker='x', c='yellow')
plt.scatter(sample_data[cluster==1, 0], sample_data[cluster==1, 1], marker='x', c='orange')
plt.scatter(sample_data[cluster==2, 0], sample_data[cluster==2, 1], marker='x', c='green')
plt.scatter(kmeans_model.cluster_centers_[:, 0], kmeans_model.cluster_centers_[:, 1], marker='o', c='red')
plt.title('Clustered Iris Dataset using sklearn Kmeans Alg')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```
